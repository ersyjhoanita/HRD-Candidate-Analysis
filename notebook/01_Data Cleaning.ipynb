## Data Cleaning & Preprocessing
Dataset: IBM HR Analytics Employee Attrition

# Tujuan Tahap Ini
Menyiapkan data agar siap dianalisis & dimodelkan

import library dan load dataset

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler

# Load dataset
df = pd.read_csv("../dataset/HR-Employee-Attrition.csv")

# Lihat 5 data teratas
df.head()
```

Cek Struktur & Missing Values

```python
# Informasi dataset
df.info()
# Cek missing values
df.isnull().sum().sort_values(ascending=False)

```

Drop Kolom Tidak Informatif

```python
drop_cols = [
    "EmployeeCount",
    "EmployeeNumber",
    "StandardHours",
    "Over18"
]

df.drop(columns=drop_cols, inplace=True)
```

Encoding Target Variable
Target: Attrition → binary classification

```python
df["Attrition"] = df["Attrition"].map({"Yes": 1, "No": 0})
```

Pisahkan Fitur & Target

```python
X = df.drop("Attrition", axis=1)
y = df["Attrition"]
```

memisahkan Kolom Numerik & Kategorikal

```python
categorical_cols = X.select_dtypes(include="object").columns
numerical_cols = X.select_dtypes(include=["int64", "float64"]).columns
```

Encoding Data Kategorikal
menggunakan One-Hot Encoding, karena tidak memberi makna urutan palsu, cocok untuk Logistic Regression, dan Feature importance tetap bisa dibaca

```python
df = pd.get_dummies(
    df,
    columns=categorical_cols,
    drop_first=True
)
```

Feature Scaling (Numerical Only)
ini penting untuk 
- Logistic Regression
- Interpretasi koefisien
- Model berbasis jarak

```python
scaler = StandardScaler()
X[numerical_cols] = scaler.fit_transform(X[numerical_cols])


# cek data set
df.head()
df.shape
df.describe()
```
